```python
#5
import matplotlib.pyplot as plt
import pandas as pd

df = pd.read_csv("AirPassengers.csv")
df['Month'] = pd.to_datetime(df['Month'])
df['Year'] = df['Month'].dt.year
df['Month_Num'] = df['Month'].dt.month   #extracts the month number and stores it to new column named month_num

# Plot data for each year
plt.figure(figsize=(10, 5))   #width,height
for year in df['Year'].unique():
    yearly_data = df[df['Year'] == year]
    plt.plot(yearly_data['Month_Num'], yearly_data['#Passengers'], label=str(year))

plt.xlabel("Month")
plt.ylabel("Passengers")
plt.title("Air Passengers Trend")
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd

df = pd.read_csv("AirPassengers.csv")

# Convert 'Month' column to datetime format
df['Month'] = pd.to_datetime(df['Month'])
df['Year'] = df['Month'].dt.year

# Group data by year and get total passengers per year
yearly_data = df.groupby('Year')['#Passengers'].sum()

# Plot two lines with different styles, widths, and colors
plt.figure(figsize=(10, 5))
plt.plot(yearly_data.index, yearly_data, label="Total Passengers", color='b', linewidth=2, linestyle='-')
plt.plot(yearly_data.index, yearly_data.rolling(3).mean(), label="3-Year Moving Avg", color='r', linewidth=3, linestyle='--')

plt.xlabel("Year")
plt.ylabel("Total Passengers")
plt.title("Yearly Airline Passengers Trend")
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd


df = pd.read_csv("AirPassengers.csv")

df['Month'] = pd.to_datetime(df['Month'])
df['Year'] = df['Month'].dt.year

yearly_data = df.groupby('Year')['#Passengers'].sum()

plt.figure(figsize=(15, 10))

# Line Chart (Yearly Trend)
plt.subplot(2, 3, 1)  # 2 rows, 3 columns, Position 1
plt.plot(yearly_data.index, yearly_data, 'bo-', label="Total Passengers")
plt.xlabel("Year")
plt.ylabel("Total Passengers")
plt.title("Yearly Airline Passengers Trend")
plt.legend()

# Bar Chart (Yearly Passenger Count)
plt.subplot(2, 3, 2)  # Position 2
plt.bar(yearly_data.index, yearly_data, color='orange')
plt.xlabel("Year")
plt.ylabel("Total Passengers")
plt.title("Yearly Passenger Count")

# Scatter Plot (Monthly Passenger Trend)
plt.subplot(2, 3, 3)  # Position 3
plt.scatter(df['Month'], df['#Passengers'], color='green', alpha=0.5)
plt.xlabel("Year")
plt.ylabel("Passengers")
plt.title("Monthly Passenger Trend")

# Histogram (Passenger Distribution)
plt.subplot(2, 3, 4)  # Position 4
plt.hist(df['#Passengers'], bins=10, color='purple', alpha=0.7)
plt.xlabel("Passenger Count")
plt.ylabel("Frequency")
plt.title("Passenger Distribution")

# Box Plot (Passenger Variation)
plt.subplot(2, 3, 5)  # Position 5
plt.boxplot(df['#Passengers'])
plt.ylabel("Passengers")
plt.title("Passenger Count Variability")

# Adjust layout and show plot
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import datetime

# Data
dates = ['03-10-16', '04-10-16', '05-10-16', '06-10-16', '07-10-16']
closing_values = [772.559998, 776.429993, 776.469971, 776.859985, 775.080017]
opening_values = [775.0, 780.0, 778.0, 779.5, 776.0]

# Convert string dates to datetime format
dates = [datetime.datetime.strptime(date, "%d-%m-%y") for date in dates]

# Create multiple subplots
fig, axs = plt.subplots(2, 1, figsize=(8, 10))

# Plot Closing Prices
axs[0].plot(dates, closing_values, marker='o', linestyle='-', color='r', label='Closing Price')
axs[0].set_title('Alphabet Inc. Closing Prices (Oct 3 - Oct 7, 2016)')
axs[0].set_ylabel('Price (USD)')
axs[0].legend()
axs[0].grid(True, which='major', linestyle='-', linewidth=0.8, color='blue')
axs[0].minorticks_on()
axs[0].grid(True, which='minor', linestyle=':', linewidth=0.5, color='gray')
axs[0].set_xticks([])
axs[0].set_yticks([])

# Plot Opening Prices
axs[1].plot(dates, opening_values, marker='o', linestyle='-', color='g', label='Opening Price')
axs[1].set_title('Alphabet Inc. Opening Prices (Oct 3 - Oct 7, 2016)')
axs[1].set_xlabel('Date')
axs[1].set_ylabel('Price (USD)')
axs[1].legend()
axs[1].grid(True, which='major', linestyle='-', linewidth=0.8, color='blue')
axs[1].minorticks_on()
axs[1].grid(True, which='minor', linestyle=':', linewidth=0.5, color='gray')
axs[1].set_xticks([])
axs[1].set_yticks([])

# Adjust layout and show plot
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt

mean_velocity = [0.2474, 0.1235, 0.1737, 0.1824]
std_deviation = [0.3314, 0.2278, 0.2836, 0.2645]
categories = ['A', 'B', 'C', 'D']

# Create Bar Plot with Error Bars
plt.figure(figsize=(8, 5))
bars = plt.bar(categories, mean_velocity, yerr=std_deviation, capsize=5, color='skyblue', edgecolor='blue')

# Attach text labels above each bar displaying the mean velocity (rounded to integer)
for bar in bars:
    height = bar.get_height()
    # Adjust the text position to be slightly above the bar
    plt.text(bar.get_x() + bar.get_width() / 2, height + 0.02, f'{int(height)}', ha='center', va='bottom')

plt.xlabel('Categories')
plt.ylabel('Mean Velocity')
plt.title('Bar Plot with Error Bars and Text Labels')

plt.show()
```
